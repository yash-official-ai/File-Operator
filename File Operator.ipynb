```python
from datetime import datetime
import os

class JournalManager:

    def __init__(self):
        self.file = "journal.txt"

    def add_entry(self):
        entry = input("Enter your journal entry: ")

        with open(self.file, "a") as f:
            time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            f.write(f"[{time}] {entry}\n")

        print("Entry added successfully!")

    def view_entries(self):
        try:
            with open(self.file, "r") as f:
                print("\nYour Journal Entries:\n")
                print(f.read())

        except FileNotFoundError:
            print("Error: The journal file does not exist.")

    def search_entry(self):
        try:
            word = input("Enter keyword to search: ")
            found = False

            with open(self.file, "r") as f:
                for line in f:
                    if word.lower() in line.lower():
                        print(line)
                        found = True

            if not found:
                print("No matching entry found.")

        except FileNotFoundError:
            print("Error: The journal file does not exist.")

    def delete_entries(self):
        if os.path.exists(self.file):
            choice = input("Delete all entries? (yes/no): ")
            
            if choice.lower() == "yes":
                os.remove(self.file)
                print("All journal entries have been deleted. ")
        else:
            print("No journal entries to delete.")

j = JournalManager()

while True:
    print("\n--- Personal Journal Manager ---")
    print("1. Add Entry")
    print("2. View Entries")
    print("3. Search Entry")
    print("4. Delete Entries")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        j.add_entry()

    elif choice == "2":
        j.view_entries()

    elif choice == "3":
        j.search_entry()

    elif choice == "4":
        j.delete_entries()

    elif choice == "5":
        print("Thank you for using Personal Journal Manager.")
        break

    else:
        print("Invalid option!")
    
            
```
